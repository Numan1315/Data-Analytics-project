```python
for i in range(len(df['rate'])):
    if(df['rate'][i]=="NEW" and len(df['reviews_list'][i])<3):
        df['rate'][i]=float(df['rate'][i].replace("NEW",'0.0'))
    if(df['rate'][i]=="NEW" and len(df['reviews_list'][i])>2):
        df['rate'][i]=np.nan
    elif(df['rate'][i]=="-" and len(df['reviews_list'][i])==2):
        df['rate'][i]=float(df['rate'][i].replace("-",'0.0'))
    elif(df['rate'][i]=="-" and len(df['reviews_list'][i])>2):
        df['rate'][i]=np.nan
    elif(type(df['rate'][i])==str):
        df['rate'][i]=float(df['rate'][i][0:3])
    else:
        continue  
        
    
        
```

### Cleaning of reviews i.e removing punctuations,stop words,urls,etc..

```python
import string
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["reviews"] = df["reviews"].apply(lambda text: remove_punctuation(text))
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["reviews"] = df["reviews"].apply(lambda text: remove_stopwords(text))
```

```python
import re
pat = r'[0-9]'
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    k = url_pattern.sub(r'', text)
    return  re.sub(pat,'',k)
df["reviews"] = df["reviews"].apply(lambda text: remove_urls(text))
```

```python
from statistics import mean
from decimal import Decimal
for i in range(len(df)):
    #print(i)
    if(len(df['list_review_rate'][i]) !=0):
        avg = round(mean(df['list_review_rate'][i]) , 2)
    if(df['votes'][i] > 100 and len(df['list_review_rate'][i]) !=0): 
        vote = df['votes'][i]
        y = len(df['list_review_rate'][i])
        x = round(Decimal(vote.item()) / Decimal(y*2) , 2)
        x = round(Decimal(vote.item()) / Decimal(x) , 2)
        z = x +y
        w1 = round(Decimal(x)/ Decimal(z) , 2)
        w2 = round(Decimal(y)/ Decimal(z) , 2)
        df['final_rate'][i] = round(w1 * Decimal(df['rate'][i].item()) + w2 * Decimal(avg) ,1)
    else:
        if(df['votes'][i] < len(df['list_review_rate'][i])):
            df['final_rate'][i] = round(avg,1)
        else:
            df['final_rate'][i] = round(0.5 * (df['rate'][i]) + 0.5 * avg,1)
 
```
