## Handling missing values for rate column

```python
type(df['rate'][0])
```

```python
df['rate'].unique()
```

```python
dict(df['rate'].value_counts())
```

```python
for i in range(len(df['rate'])):
    if(df['rate'][i]=="NEW" and len(df['reviews_list'][i])<3):
        df['rate'][i]=float(df['rate'][i].replace("NEW",'0.0'))
    if(df['rate'][i]=="NEW" and len(df['reviews_list'][i])>2):
        df['rate'][i]=np.nan
    elif(df['rate'][i]=="-" and len(df['reviews_list'][i])==2):
        df['rate'][i]=float(df['rate'][i].replace("-",'0.0'))
    elif(df['rate'][i]=="-" and len(df['reviews_list'][i])>2):
        df['rate'][i]=np.nan
    elif(type(df['rate'][i])==str):
        df['rate'][i]=float(df['rate'][i][0:3])
    else:
        continue  
        
    
        
```

```python
for i in range(len(df['rate'])):
    l=[]
    x=0
    y=0
    if(np.isnan(df['rate'][i]) and len(df['reviews_list'][i])>2):
        while(x<len(df['reviews_list'][i])):
            x=df['reviews_list'][i].find('Rated ',x)
            y=df['reviews_list'][i].find(')',x)
            if(x==-1):
                break;
            l.append(float(df['reviews_list'][i][x+6:x+9]))
            x=y
        df['rate'][i]=float(format(sum(l)/len(l),".1f"))
df=df.dropna(subset=['rate'])
df=df.reset_index(drop=True)
```

```python
df['rate']=df['rate'].astype('float')
```

Handling the missing values present in the dish_liked column

```python
for i in range(len(df['dish_liked'])): 
    dish=[]
    if(type(df['dish_liked'][i])==float):
        continue
    for t in df['dish_liked'][i].split(','):
            t=t.strip() 
            dish.append(t)
    df['dish_liked'][i]=dish
```

```python

for i in range(len(df['menu_item'])):
    menu=[]
    df['menu_item'][i]=df['menu_item'][i].replace('[',' ').strip()
    df['menu_item'][i]=df['menu_item'][i].replace(']',' ').strip()
    df['menu_item'][i]=df['menu_item'][i].replace('\'',' ').strip()

    if(len(df['menu_item'][i])>0):
        for t in df['menu_item'][i].split(','):
            t=str(t.strip()) 
            menu.append(t)
    df['menu_item'][i]=menu
```

```python
df.dish_liked.isnull().sum()
```

```python
import random
for i in range(len(df['dish_liked'])):
    if(type(df['dish_liked'][i])==float):
        if(len(df['menu_item'][i])==0):
            continue
        elif(len(df['menu_item'][i]) > 5):
            df['dish_liked'][i]=random.sample(df['menu_item'][i] , 5)
        else:
            df['dish_liked'][i] = df['menu_item'][i]


            
            
            

```

```python
for i in range(len(df['dish_liked'])):
    if(type(df['dish_liked'][i])==float):
        x=df[df['name']==df['name'][i]]
        #print(len(x))
        if(len(x)>1):
            y=[]
            for a in x['dish_liked']:
                if(type(a)==float):
                    continue
                else:
                    y=y+list(set(a))
            y=list(set(y))
            if(len(y) > 5):
                df['dish_liked'][i]=random.sample(y , 5)
            else:
                df['dish_liked'][i] = y
```

Cleaning of the reviews column since it had some reviews which were not in English language

```python
def clear_text(t):
    return ' '.join([i[1] for i in t]).encode('utf8').decode('ascii',errors='replace')
```

```python
import ast
df.reviews_list = df.reviews_list.apply(lambda x: ast.literal_eval(x))
type(df.reviews_list[0])
```

```python
df['reviews'] = df.reviews_list.apply(lambda x: clear_text(x))
```

```python
df['reviews'] = df.reviews.apply(lambda x : x.replace("RATED\n  ",'').replace('?','').replace('�','').replace('\n','').replace('.',' ').strip().lower())
```
